```python
import gym
import numpy as np
import cv2
import pickle as pkl
```

```python
cliffenv=gym.make("CliffWalking-v0")
```

```python
ALPHA = 0.1  # Learning ra = 0.9  # Discount factor
EPSILON = 0.1  # Exploration rate
EPISODES = 500
GAMMA=0.9
```

```python
q_table=np.zeros(shape=(48,4))
```

```python
def policy(state, explore=0.0):
    state = int(state)
    action = int(np.argmax(q_table[state]))
    if np.random.random() <= explore:
        action = int(np.random.randint(low=0, high=1, size=1))
    return action
```

```python
!pip install --upgrade gym
!pip install numpy==1.23.5
from tqdm import tqdm  # ADD THIS IMPORT

# Update the unpacking of the step method's return value
for episode in tqdm(range(EPISODES), desc="Training Progress"):
  done=False
  total_reward=0
  episode_length=0

  state, info = cliffenv.reset()
  while not done:
    action=policy(state,EPSILON)
    # Unpack 5 values instead of 4 from the step method
    next_state, reward, terminated, truncated, info = cliffenv.step(action)
    # In newer gym versions, 'done' is represented by 'terminated' or 'truncated'
    done = terminated or truncated

    next_action=policy(next_state)

    q_table[state, action] += ALPHA * (reward + GAMMA * q_table[next_state, next_action] - q_table[state, action])
    state=next_state
    total_reward+=reward
    episode_length+=1

  print("episode",episode,"length:",episode_length,"total_-reward",total_reward)

cliffenv.close()
pkl.dump(q_table,open("q_learning_q_table.pkl","wb"))
```
